```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'Resources/73515083_102017_2041_airline_delay_causes.csv'
raw_pd = pd.read_csv(file, encoding="iso-8859-1", low_memory=False)
raw_pd.head()
```

```python
raw_pd['Total Percentage Delayed'] = raw_pd['arr_del15']/raw_pd['arr_flights']* 100
raw_pd['Carrier Delay'] = raw_pd['carrier_ct']/raw_pd['arr_flights']* 100
raw_pd['Weather Delay'] = raw_pd['weather_ct']/raw_pd['arr_flights']* 100
```

```python

refine_df = raw_pd[['airport','TPercentage Delayed','Carrier Delay' ]]
```
