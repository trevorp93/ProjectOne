```python
import unicodecsv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from scipy import stats
from scipy.stats import norm

pd.options.display.float_format = '{:.0f}'.format
```

```python
file = '../Resources/reformed_data.csv'
file2 = '../Resources/73515083_102017_2041_airline_delay_causes.csv'
df = pd.read_csv(file, encoding="iso-8859-1", low_memory=False)
df2 = pd.read_csv(file2, encoding="iso-8859-1", low_memory=False)


```

```python
df['Total Percentage Delayed'] = df['Arrival_Delay']/df['arr_flights']* 100
df['Carrier Delay %'] = df['Carrier_Delay']/df['arr_flights']* 100
df['Weather Delay %'] = df['Weather_Delay']/df['arr_flights']* 100
df['NAS %'] = df['NAS_Delay']/df['arr_flights']* 100
df['Late %'] = df['Late_Aircraft_Delay']/df['arr_flights']* 100
df['Security %'] = df['Security_Delay']/df['arr_flights']* 100
df['Cancelled %']= df['Cancelled']/(df['arr_flights']+df['Cancelled'])* 100
```

```python
df3= df2[['carrier_name','weather_delay']]
Cdf = df[['carrier_name','Total Percentage Delayed','Carrier Delay %','Weather Delay %','Cancelled %','NAS %','Late %','Security %']]

```

```python
df4=df3.groupby('carrier_name').mean().reset_index()
ax = sns.barplot(x="weather_delay", y="carrier_name", data=df4)
plt.show()
```

```python



CarrierD = df['Carrier_Delay'].mean()
WeatherD = df['Weather_Delay'].mean()
NASD = df['NAS_Delay'].mean()
SecurityD = df['Security_Delay'].mean()
LateD = df['Late_Aircraft_Delay'].mean()             

```

```python

A=Cdf.groupby(['carrier_name']).mean().reset_index()
```
